```python
import os
import numpy as np
import pandas as pd
```

```python
master = pd.read_excel("../data/meta/Master_List.xlsx")
master.drop(columns=['Sample_body_site', 'Type', 'urls'], inplace=True)
#master.drop_duplicates(subset=['Subject_ID', 'Visit_number', 'Visit_ID', 'Age', 'Sex', 'Diagnosis', 'Race'], inplace=True)
unique_master = master.drop_duplicates(subset=['Subject_ID', 'Visit_number', 'Visit_ID', 'Age', 'Sex', 'Diagnosis', 'Race'])
```

```python
def visit_stage(visit):
    if visit >= 4 and visit <=9:
        return "stage 1"
    elif visit >= 11 and visit <= 16:
        return "stage 2"
    elif visit >= 18 and visit <= 23:
        return "stage 3"
    elif visit >= 25 and visit <= 30:
        return "stage 4"
    else:
        return "unknown"

def age_group(age):
    if age <= 18:
        return 1
    elif age > 18 and age <= 40:
        return 2
    elif age > 40:
        return 3
```

```python
unique_master.loc[:, ['age_group']] = unique_master['Age'].apply(lambda age: age_group(age))
unique_master.loc[:, ['visit_stage']] = unique_master['Visit_number'].apply(lambda visit: visit_stage(visit))
unique_master
```

```python
unique_master[:, ['age_group']] = unique_master.age_group.astype('category')
unique_master[:, ['visit_stage']] = unique_master.visit_stage.astype('category')
```
