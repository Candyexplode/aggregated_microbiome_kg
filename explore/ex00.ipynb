```python
tax_dir = "./data/csv/tax"
tax_files = os.listdir(tax_dir)
tax_files = [os.path.join(tax_dir, f) for f in tax_files]
df_tax = pd.concat(map(pd.read_csv, tax_files), ignore_index=True)
df_tax.head()
```

```python
len(df_tax['Visit_name'].unique())
```

```python
master_no_dup = master.drop_duplicates(subset=['Subject_ID', 'Visit_number', 'Visit_ID', 'Age', 'Sex', 'Diagnosis', 'Race'])
master_no_dup.drop(columns=['Sample_body_site', 'Type', 'urls'], inplace=True)
master_no_dup
```

```python
genus_abundances = df_tax.loc[:,["Visit_name", 'Genus', 'Metaphlan2_Analysis']]
genus_abundances.rename(columns={'Visit_name': 'Visit_ID'}, inplace=True)
```

```python
df_master_tax = pd.merge(master_no_dup, genus_abundances, on='Visit_ID')
df_master_tax
```

```python
genus_abundances.reset_index(inplace=True)
genus_abundances
```

```python
genus_abundances.rename(columns={'index': 'idx'}, inplace=True)
genus_abundances
```

```python
genus = genus_abundances.pivot(index='idx', columns='Genus', values='Metaphlan2_Analysis')
genus
```

```python
genus['Visit_ID'] = genus_abundances['Visit_ID']
genus
```

```python
master_no_dup
```

```python
df_master_tax = pd.merge(master_no_dup, genus, on='Visit_ID')
```

```python
df_master_tax
```

```python
df_master_tax = df_master_tax.fillna(0.0)
df_master_tax
```

```python
df_master_tax.to_csv("./data/agg/master_tax.csv")
```
