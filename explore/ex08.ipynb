```python
genus_abundances = tax.loc[:,["Visit_name", 'Genus', 'Metaphlan2_Analysis']]
genus_abundances.reset_index(inplace=True)
genus_abundances.rename(columns={'index': 'idx', 'Visit_name': 'Visit_ID'}, inplace=True)
genus = genus_abundances.pivot(index='idx', columns='Genus', values='Metaphlan2_Analysis')
genus['Visit_ID'] = genus_abundances['Visit_ID']
group_genus = genus.groupby('Visit_ID').sum()
```

```python
def visit_stage(visit):
    if visit >= 4 and visit <=9:
        return "stage 1"
    elif visit >= 11 and visit <= 16:
        return "stage 2"
    elif visit >= 18 and visit <= 23:
        return "stage 3"
    elif visit >= 25 and visit <= 30:
        return "stage 4"
    else:
        return "unknown"

def age_group(age):
    if age <= 18:
        return 1
    elif age > 18 and age <= 40:
        return 2
    elif age > 40:
        return 3
```

```python
unique_master.loc[:, ['age_group']] = unique_master['Age'].apply(lambda age: age_group(age))
unique_master.loc[:, ['visit_stage']] = unique_master['Visit_number'].apply(lambda visit: visit_stage(visit))
unique_master
```

```python
unique_master.loc[:, ['age_group']] = unique_master.age_group.astype('category')
unique_master.loc[:, ['visit_stage']] = unique_master.visit_stage.astype('category')
```
