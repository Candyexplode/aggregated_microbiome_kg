```python
import os
import numpy as np
import pandas as pd
```

```python
master = pd.read_excel("../data/meta/Master_List.xlsx")
master.drop(columns=['Sample_body_site', 'Type', 'urls'], inplace=True)
unique_master = master.drop_duplicates(subset=['Subject_ID', 'Visit_number', 'Visit_ID', 'Age', 'Sex', 'Diagnosis', 'Race'])
```

```python
dir = "../data/genus_cleaned/ecs/"
files = os.listdir(dir)
files = [os.path.join(dir, f) for f in files]
ecs = pd.concat(map(pd.read_csv, files), ignore_index=True)
```

```python
dir = "../data/genus_cleaned/pwy/"
files = os.listdir(dir)
files = [os.path.join(dir, f) for f in files]
pwy = pd.concat(map(pd.read_csv, files), ignore_index=True)
```

```python
dir = "../data/genus_cleaned/tax/"
files = os.listdir(dir)
files = [os.path.join(dir, f) for f in files]
tax = pd.concat(map(pd.read_csv, files), ignore_index=True)
```

```python
genus_abundances = tax.loc[:,["Visit_name", 'Genus', 'Metaphlan2_Analysis']]
genus_abundances.reset_index(inplace=True)
genus_abundances.rename(columns={'index': 'idx', 'Visit_name': 'Visit_ID'}, inplace=True)
genus = genus_abundances.pivot(index='idx', columns='Genus', values='Metaphlan2_Analysis')
genus['Visit_ID'] = genus_abundances['Visit_ID']
group_genus = genus.groupby('Visit_ID').sum()
```

```python
master_tax = pd.merge(unique_master, group_genus, on='Visit_ID')
master_tax = master_tax.fillna(0.0)
master_tax.to_csv(os.path.join('../data/agg/', "master_tax_cleaned.csv"))
```

```python
# master_tax[master_tax.filter(like='g_').abs().gt(1.0).any()]
genus_columns = master_tax.columns[7:]
outliers = []
for col in genus_columns:
    outliers.append(master_tax[master_tax[col] > 100.0])
```

```python
outs = [out for out in outliers if len(out) > 0]
result = pd.concat(outs)
```

```python
result
```

```python
cleaned_master_tax = master_tax[~master_tax.index.isin(result.index)]
```
